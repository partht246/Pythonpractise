```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)#categorical data based of a function #mean by default
#average total bill of males and females
```

```python
#sns.barplot(x='sex',y='total_bill',data=tips,estimator=)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips) #more males in dataset then females
```

```python
sns.boxplot(x='day',y='total_bill',data=tips) #each figure is quarter distribution graph
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips) #density
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter = True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter = True,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar') #creates all kind of plot through kind
```
