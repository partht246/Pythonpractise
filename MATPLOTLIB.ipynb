```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
#FUNCTIONAL METHOD
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Graph')
#labels
```

```python
plt.xlabel('X label')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'c')

plt.subplot(1,2,2)
plt.plot(y,x,'r')
#multiple plots on same canvas
```

```python
#object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y) ##more control of the plot
axes.set_xlabel('X label')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #left se 20% se start hoga, 50%bottom se upar start hoga, 40%width wide hoga ,30%length hoga
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])
axes2 = fig.add_axes([0.6,0.3,0.2,0.3]) #60% to the left,30% to the top(from the bottom),20% canvas size width, 30% canvas size; reference (from) is from 0,0 
```
